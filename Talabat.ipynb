## Results obtained from Exploratory Data Analysis (EDA)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data1 = pd.read_csv('dim_order_details.csv')

data1.info()
#127,079 orders
```

```python
data2 = pd.read_csv('fct_orders.csv')
data3 = pd.read_csv('dim_countries.csv')
data4 = pd.read_csv('dim_chains.csv')

most_data = data1.merge(data2).merge(data3)
all_data = pd.merge(most_data, data4, how = 'outer')

all_data['chain_name'].fillna('Not Available', inplace = True)
all_data['order_date'] = pd.to_datetime(all_data.order_date, format='%Y-%m-%d %H:%M:%S.%f')
all_data.amt_discount_eur = all_data.amt_discount_eur.round(2)
all_data.amt_basket_euro = all_data.amt_basket_euro.round(2)
all_data.rename(columns = {'orderid': 'order_id', 'countryid':'country_id', 'chainid': 'chain_id', 'customerid': 'customer_id'},
               inplace = True)
all_data = all_data.astype({'customer_id': 'str', 'country_id': 'str', 'chain_id': 'str'})


all_data.insert(loc=2, column='year_month', value=all_data['order_date'].map(lambda x: 100*x.year + x.month))
all_data.insert(loc=2, column='year', value=all_data.order_date.dt.year)
all_data.insert(loc=3, column='month', value=all_data.order_date.dt.month)
all_data.insert(loc=4, column='day', value=(all_data.order_date.dt.dayofweek)+1)
all_data.insert(loc=5, column='hour', value=all_data.order_date.dt.hour)


all_data.head()
```

```python
datatoexcel = pd.ExcelWriter('all_data.xlsx', engine='xlsxwriter')

all_data.to_excel(datatoexcel, sheet_name = 'Sheet1', index = False)

datatoexcel.save()
```
